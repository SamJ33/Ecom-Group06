```python
import re

def extract_keywords(text):
    if pd.isna(text):
        return ""
    
    # Lowercase
    text = text.lower()
    
    # Remove numbers
    text = re.sub(r"\d+", "", text)
    
    # Remove punctuation
    text = re.sub(r"[^\w\s]", "", text)
    
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    
    # Tokenize and remove stopwords
    words = text.split()
    keywords = [w for w in words if w not in portuguese_stopwords and len(w) > 2]
    
    return " ".join(keywords)

df_nlp_p["keywords_nlp"] = df_nlp_p["text_for_nlp"].apply(extract_keywords)
```

```python
# Flatten all keyword lists into a single list
all_keywords = [word for sublist in df_nlp_p['keywords_nlp'] for word in sublist]

# Keep only unique words to minimize translation calls
unique_keywords = list(set(all_keywords))
print(f"Unique words to translate: {len(unique_keywords)}")
```

```python
from deep_translator import GoogleTranslator

# Extract unique keywords
unique_keywords = list(set([word for sublist in df_nlp_p['keywords_nlp'] for word in sublist]))

# Translate keywords to English
translator = GoogleTranslator(source='pt', target='en')
translated_dict = {word: translator.translate(word) for word in unique_keywords}

# Map back to dataframe
df_nlp_p['keywords_nlp_en'] = df_nlp_p['keywords_nlp'].apply(
    lambda kw_list: [translated_dict[word] for word in kw_list]
)
```
