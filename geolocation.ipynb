```python
df_geo_clean = (
    df_geo
    .groupby("geolocation_zip_code_prefix")
    .agg({
        "geolocation_lat": "mean",
        "geolocation_lng": "mean",
        "geolocation_city": "first",
        "geolocation_state": "first"
    })
    .reset_index()
)
```

```python
df_geo["geolocation_state"].value_counts()
```

```python
df_geo_clean = df_geo.drop(
    columns=["geolocation_city", "geolocation_lat", "geolocation_lng"],
    errors="ignore"
)
```

```python
print(df_geo_clean.head())
print(df_geo_clean.columns)
```

```python


df_geo_clean["geolocation_state"] = (
    df_geo_clean["geolocation_state"]
    .str.strip()
    .str.upper()
)
```
