```python
import pandas as pd
import numpy as np
import re
from collections import Counter

# Load data

df = pd.read_csv(r"C:\Users\user\Desktop\Ecom\data\raw\olist_customers_dataset.csv")
df.head()


# Comprehensive cleaning function
def clean_customer_data(df):
    """
    Comprehensive cleaning function for customer data
    """
    # Make a copy to avoid warnings
    df_clean = df.copy()
    
    # 1. Clean string columns - remove quotes and extra spaces
    string_columns = ['customer_id', 'customer_unique_id', 'customer_city', 'customer_state']
    for col in string_columns:
        df_clean[col] = df_clean[col].astype(str).str.strip('"').str.strip()
    
    # 2. Clean zip code
    df_clean['customer_zip_code_prefix'] = df_clean['customer_zip_code_prefix'].astype(str).str.strip('"')
    df_clean['customer_zip_code_prefix'] = df_clean['customer_zip_code_prefix'].str.zfill(5)
    
    # 3. Standardize city names (proper case)
    df_clean['customer_city'] = df_clean['customer_city'].str.lower().str.title()
    
    # 4. Standardize state codes (uppercase)
    df_clean['customer_state'] = df_clean['customer_state'].str.upper()
    
    # 5. Remove duplicates keeping first occurrence
    initial_rows = len(df_clean)
    df_clean = df_clean.drop_duplicates(subset=['customer_id'], keep='first')
    print(f"Removed {initial_rows - len(df_clean)} duplicate customer_ids")
    
    # 6. Handle missing values
    missing_before = df_clean.isnull().sum()
    # If any missing values, we can fill or drop based on business rules
    # df_clean = df_clean.dropna()  # Uncomment if you want to drop missing
    
    return df_clean

# Apply cleaning
df_cleaned = clean_customer_data(df)

# Validation checks
print("\n" + "="*60)
print("VALIDATION CHECKS")
print("="*60)

# Check 1: All customer_ids should be unique
assert df_cleaned['customer_id'].is_unique, "customer_id column has duplicates!"

# Check 2: All zip codes should be 5 digits
invalid_zips = df_cleaned[~df_cleaned['customer_zip_code_prefix'].str.match(r'^\d{5}$')]
if len(invalid_zips) > 0:
    print(f"Warning: Found {len(invalid_zips)} invalid zip codes")
else:
    print("✓ All zip codes are valid 5-digit numbers")

# Check 3: State codes should be valid Brazilian states
valid_states = ['AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MT', 'MS', 
                'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 'RS', 'RO', 'RR', 'SC', 
                'SP', 'SE', 'TO']
invalid_states = df_cleaned[~df_cleaned['customer_state'].isin(valid_states)]
if len(invalid_states) > 0:
    print(f"Warning: Found {len(invalid_states)} invalid state codes")
    print(invalid_states['customer_state'].value_counts())
else:
    print("✓ All state codes are valid")

# Check 4: No null values
null_counts = df_cleaned.isnull().sum()
if null_counts.sum() > 0:
    print(f"Warning: Found null values:\n{null_counts[null_counts > 0]}")
else:
    print("✓ No null values found")

# Summary statistics
print("\n" + "="*60)
print("CLEANED DATA SUMMARY STATISTICS")
print("="*60)
print(f"Total records: {len(df_cleaned):,}")
print(f"Unique customers: {df_cleaned['customer_unique_id'].nunique():,}")
print(f"States covered: {df_cleaned['customer_state'].nunique()}")
print(f"Cities covered: {df_cleaned['customer_city'].nunique():,}")

# Top states and cities
print("\nTop 5 States:")
print(df_cleaned['customer_state'].value_counts().head())

print("\nTop 10 Cities:")
print(df_cleaned['customer_city'].value_counts().head(10))

# Save cleaned data
df_cleaned.to_csv('olist_customers_cleaned_final.csv', index=False)
print("\n✓ Cleaned data saved to 'olist_customers_cleaned_final.csv'")

# Display first few rows of cleaned data
print("\nFirst 10 rows of cleaned data:")
print(df_cleaned.head(10))
```
