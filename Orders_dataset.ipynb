```python
df_order=df_order[~((df_order['order_status']=='delivered') & (df_order['order_delivered_carrier_date'].isna()))]
```

```python
# To understand behavior:
def delivery_behavior(row):
     if pd.notna(row['order_delivered_customer_date']):
         return 'Delivered'
         
     if row['order_status'] == 'canceled' and pd.notna(row['order_delivered_carrier_date']):
         return 'Canceled_After_Shipping'
         
     if row['order_status'] == 'canceled':
        return 'Canceled_Before_Shipping'
     
     if row['order_status'] == 'shipped':
        return 'In_Transit'
         
     return 'Not_FulFilled'

df_order['Delivery_Behavior']=df_order.apply(delivery_behavior, axis=1)
```
