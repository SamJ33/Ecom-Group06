## NLP and Sentiment Analysis
Using a multilingual transformer model to translate from Portugese to English then deduce the sentiment analysis score 

```python
import torch
```

```python
from transformers import MarianMTModel, MarianTokenizer
from tqdm import tqdm
```

```python
model_name = "Helsinki-NLP/opus-mt-ROMANCE-en"
token="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"
tokenizer = MarianTokenizer.from_pretrained(model_name,token=token)
model= MarianMTModel.from_pretrained(model_name, token=token)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

### Test a Sample n=1000

```python
df_test = df_nlp.head(1000).copy()

def translate_batch(texts, batch_size=32):
    translated = []

    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i:i+batch_size]
        
        inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True, max_length=512)
        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = model.generate(**inputs)

        decoded = tokenizer.batch_decode(outputs, skip_special_tokens=True)
        translated.extend(decoded)

    return translated
```

```python
import time
start = time.time()

_ = translate_batch(df_test["text_for_nlp"].tolist())

print("Time:", time.time() - start)
```
