# Data Cleaning

#### order_id column requires dictionaire remapping

```python
unique_orders=df_order_payment['order_id'].unique()
```

```python
unique_orders
```

```python
#Creating a dictionaire for the mapping
order_id_map={}
new_id=1

for old_id in unique_orders:
    order_id_map[old_id]=new_id
    new_id+=1
```

```python
order_id_map
```

```python
df_order_payment['order_id']=df_order_payment['order_id'].map(order_id_map)
```

```python
df_order_payment
```

## Payment Sequential Segmentation
The segments are divided into 2:


Equal to 1 (majority)


Greater than 1

```python
df_order_payment['payment_segment']='>1'
df_order_payment.loc[df_order_payment['payment_sequential']==1,'payment_segment'] = '1'
```

```python
df_order_payment
```

```python
df_order_payment.drop(columns=(['payment_sequential']),inplace=True)
```

```python
df_order_payment
```

### Payment Installments
Categrize them :
- 1:Single Payment
- 2–3:	Low Installments
- 4–6:	Medium Installments
- 7+:   High Installments


```python
#Create a function to categorize each one:
def categorize_installments(x):
    if x==1:
        return '1'
    elif 2<=x<=3:
        return '2-3'
    elif 4<=x<=6:
        return '4-6'
    else:
        return '7+'

df_order_payment['installment_segment']=df_order_payment['payment_installments'].apply(categorize_installments)
```

```python
df_order_payment
```

```python
df_order_payment.drop(columns=['payment_installments'],inplace=True)
```

```python
df_order_payment
```

## Payment Type
There is a type called : not defined
It contains a transaction that has no payment value therefore it is not valuable to the study

```python
#Create a mask to remove the rows with the type ="not_defined"
mask = df_order_payment['payment_type'] != 'not_defined'
df_order_payment = df_order_payment[mask]
```

```python
df_order_payment['payment_type'].unique()
```
